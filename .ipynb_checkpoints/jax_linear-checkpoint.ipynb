```python
import jax.numpy as jnp
from jax import value_and_grad, jit
import jax
```

```python
from jax import config
config.update("jax_debug_nans", True)
```

```python
from jax import random
```

```python
a, b, c = 4, 3, 8
seed = 54
key = random.PRNGKey(seed)
x = random.normal(key, (1000, 1))
y = a*x**2 + b*x + c
```

```python
x.shape, y.shape
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x, y)
plt.show();
```

```python
def init_params(layer_sizes, key):
    params = []
    keys = random.split(key, num = len(layer_sizes))
    
    for in_features, out_features, key in zip(layer_sizes[:-1], layer_sizes[1:], keys):
        wkey, bkey = random.split(key)
        weight = random.uniform(wkey, (in_features, out_features)) * (1/in_features)**0.5 # kaiming init
        bias = random.uniform(bkey, (out_features,)) * (1/in_features)**0.5 # kaiming init
        params.append((weight, bias))
    
    return params
```

```python
layer_sizes = [1, 10, 1]
_, key = random.split(key)
params = init_params(layer_sizes, key)
```

```python
jax.tree_map(lambda x: x.shape, params)
```

```python
def predict(params, x):
    yh = jnp.dot(x, params[0][0]) + params[0][1]
    activation = jax.nn.relu(yh)
    return jnp.dot(activation, params[1][0]) + params[1][1]
```

```python
def loss(params, x, y):
    yhat = predict(params, x)
    return jnp.sqrt(jnp.mean((yhat-y)**2))
```

```python
@jit
def update(params, x, y, lr=0.01):
    loss_value, gradient = value_and_grad(loss)(params, x, y)
    params = [((w + (-lr * dw)), (b + (-lr * db))) for (w, b), (dw, db) in zip(params, gradient)]
    
    return params, loss_value
```

```python
for i in range(10000):
    params, loss_value = update(params, x, y)
    
    if i % 1000 == 0:
        print(loss_value)
```

```python
yhat = predict(params, x)
```

```python
plt.scatter(x, y)
plt.scatter(x, yhat)
plt.show();
```
